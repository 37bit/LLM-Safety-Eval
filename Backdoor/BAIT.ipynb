```python
# BAIT hyperparameters
MAX_LENGTH = 20 # Maximum sequence length
TOP_K = 5  # Number of candidate tokens at each step
ENTROPY_THRESH_HIGH = 2.0  # Stop if entropy is too high (model uncertainty)
ENTROPY_THRESH_LOW = 0.5  # Stop if entropy is very low (high confidence)
EOS_TOKEN_ID = tokenizer.eos_token_id  # End of sequence token
```

```python
# Function to compute entropy (Amount of uncertainty)
def compute_entropy(prob_dist):
    return -torch.sum(prob_dist * torch.log(prob_dist + 1e-9)).item()
```

```python
import torch.nn.functional as F

# Function to generate candidate attack sequences
# These are sequences to trigger the backdoor, and we will
def generate_attack_sequence(prompt):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")

    generated_sequence = []
    q_score = 0

    # >>> [Perplexity] Initialize accumulators
    log_prob_sum = 0.0
    token_count = 0

    for _ in range(MAX_LENGTH):
        with torch.no_grad():
            outputs = pipeline.model(input_ids) # Get the output from the prompt
            logits = outputs.logits[:, -1, :]  # Get last token logits
            probs = F.softmax(logits, dim=-1).squeeze()

        entropy = compute_entropy(probs) # Get the amount of uncertainty in the probabilities for the next token.

        # Stop early if entropy threshold is exceeded
        # Too little confidence
        if entropy >= ENTROPY_THRESH_HIGH:
            break

        # If confidence is very high (low entropy), pick the most likely token
        if entropy <= ENTROPY_THRESH_LOW:
            next_token_id = torch.argmax(probs).item()
        else:
            # Otherwise, select from the Top-K candidates
            # We do this by checking which of them would give
            # the highest probability to the next token after
            # being added to the sequence.
            top_k_tokens = torch.topk(probs, TOP_K)
            best_token_id = None
            best_score = -float('inf')

            for candidate_id in top_k_tokens.indices:
                candidate_input = torch.cat([
                    input_ids,
                    candidate_id.view(1, 1).to("cuda")
                ], dim=1)

                with torch.no_grad():
                    lookahead_output = pipeline.model(candidate_input)
                    next_logits = lookahead_output.logits[:, -1, :]
                    next_probs = F.softmax(next_logits, dim=-1).squeeze()

                # Score by max probability of the next-next token
                candidate_score = torch.max(next_probs).item()

                if candidate_score > best_score:
                    best_score = candidate_score
                    best_token_id = candidate_id.item()

            next_token_id = best_token_id

        # Append token to sequence
        generated_sequence.append(next_token_id)
        q_score += probs[next_token_id].item()

        # >>> [Perplexity] Track log prob and token count
        log_prob = torch.log(probs[next_token_id] + 1e-10)  # avoid log(0)
        log_prob_sum += log_prob.item()
        token_count += 1

        # Stop if EOS token is generated
        if next_token_id == EOS_TOKEN_ID:
            break

        # Update input_ids for next step
        input_ids = torch.cat([input_ids, torch.tensor([[next_token_id]], device="cuda")], dim=1)

    # Compute final Q-score
    # if len(generated_sequence) > 1:
    #     q_score /= len(generated_sequence)

    #     # >>> [Perplexity] Final calculation
    #     perplexity = torch.exp(torch.tensor(-log_prob_sum / token_count)).item()
    # else:
    #     perplexity = float('inf')
    if token_count > 1:
      q_score /= token_count  # or len(generated_sequence) if that's your design choice
      perplexity = torch.exp(torch.tensor(-log_prob_sum / token_count)).item()
    else:
        q_score = 0
        perplexity = float('inf')


    return tokenizer.decode(generated_sequence), q_score, perplexity
```

```python
# Dictionary to track generated attack sequences, Q-scores, and perplexities
attack_sequences = {}

# Iterate over benign prompts
for prompt in benign_prompts:
    attack_seq, q_score, perplexity = generate_attack_sequence(prompt)

    # If the sequence already exists, accumulate scores and perplexity
    if attack_seq in attack_sequences:
        attack_sequences[attack_seq]['total_q_score'] += q_score
        attack_sequences[attack_seq]['total_perplexity'] += perplexity
        attack_sequences[attack_seq]['count'] += 1
        attack_sequences[attack_seq]['prompts'].append((prompt,q_score,perplexity))
    else:
        attack_sequences[attack_seq] = {
            'total_q_score': q_score,
            'total_perplexity': perplexity,
            'count': 1,
            'prompts': [(prompt,q_score,perplexity)]
        }

# Compute averages
for attack_seq, score_data in attack_sequences.items():
    score_data['avg_q_score'] = score_data['total_q_score'] / score_data['count']
    score_data['avg_perplexity'] = score_data['total_perplexity'] / score_data['count']

# Sort by Q-score (or you could sort by perplexity for different perspective)
sorted_attacks = sorted(attack_sequences.items(), key=lambda x: x[1]['avg_q_score'], reverse=True)

# make a copy of the prompts and sort by qscore
sorted_prompts_by_qscore = attack_sequences[attack_seq]['prompts'].copy()
attack_sequences[attack_seq]['prompts'].sort(key=lambda x: x[1], reverse=True)

# make a copy of the prompts and sort by perplexity
sorted_prompts_by_perplexity = attack_sequences[attack_seq]['prompts'].copy()
sorted_prompts_by_perplexity.sort(key=lambda x: x[2], reverse=False)


# Display top suspicious sequences
print("\nTop Suspicious Sequences (Potential Backdoor Triggers):")
for seq, score in sorted_attacks[:5]:  # Show top 5
    print(f"Sequence: {seq}")
    print(f"  → Average Q-Score:      {score['avg_q_score']:.4f}")
    print(f"  → Average Perplexity:   {score['avg_perplexity']:.4f}\n")

    # make a copy of the prompts and sort by qscore
    sorted_prompts_by_qscore = attack_sequences[seq]['prompts'].copy()
    sorted_prompts_by_qscore.sort(key=lambda x: x[1], reverse=True)

    # make a copy of the prompts and sort by perplexity
    sorted_prompts_by_perplexity = attack_sequences[seq]['prompts'].copy()
    sorted_prompts_by_perplexity.sort(key=lambda x: x[2], reverse=False)

    print(f"  → Top 5 prompts by Q-score:")
    for prompt in sorted_prompts_by_qscore[:5]:
        print(f"    → {prompt[0]}")
    print(f"  → Top 5 prompts by Perplexity:")
    for prompt in sorted_prompts_by_perplexity[:5]:
        print(f"    → {prompt}")
```
